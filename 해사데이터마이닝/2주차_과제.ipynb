````python
import pandas as pd

df = pd.read_csv("https://raw.githubusercontent.com/lovedlim/bigdata_analyst_cert/main/part4/ch2/members.csv")

line = int(len(df) * 0.8)
df = df.iloc[:line]

std1 = df['f1'].std()

med = df['f1'].median()
df['f1'] = df['f1'].fillna(med)

std2 = df['f1'].std()

result = abs(std1 - std2)
print(round(result, 2))
````

````python
import pandas as pd

df = pd.read_csv("https://raw.githubusercontent.com/lovedlim/bigdata_analyst_cert/main/part4/ch2/members.csv")

std = df['age'].std()
mean = df['age'].mean()

lower = mean - (std * 1.5)
upper = mean + (std * 1.5)

cond1 = df['age'] < lower
cond2 = df['age'] > upper

print(df[cond1|cond2]['age'].sum())
````

### 1. 중앙값 (Median) 이란?

**중앙값**은 데이터를 크기순으로 나열했을 때 가장 한가운데 위치하는 값입니다.

* **왜 쓰나요?** 평균(Mean)은 너무 크거나 작은 값(이상치)에 민감하게 반응하지만, **중앙값은 이상치의 영향을 거의 받지 않아** 데이터의 중심 경향을 더 안정적으로 보여줍니다.
* **어떻게 구하나요?**
* 데이터 개수가 홀수면: 정확히 가운데 있는 값.
* 데이터 개수가 짝수면: 가운데 두 값의 산술 평균.

### 3. Colab 실습 코드 (중앙값 대체하기)

중앙값으로 결측치를 처리하는 가장 표준적인 코드 흐름입니다.

```python
import pandas as pd

# 1. 데이터 불러오기
df = pd.read_csv('data.csv')

# 2. 특정 컬럼의 중앙값 계산
median_value = df['컬럼명'].median()

# 3. 결측치(NaN)를 중앙값으로 채우기
df['컬럼명'] = df['컬럼명'].fillna(median_value)

```
