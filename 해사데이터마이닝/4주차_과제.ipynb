```python
import pandas as pd

df = pd.read_csv("https://raw.githubusercontent.com/lovedlim/bigdata_analyst_cert_v2/main/part4/ch9/loan.csv")

df["총대출액"] = df["신용대출"] + df["담보대출"]

pivot_result = df.pivot_table(
    index="지역코드",
    columns="성별",
    values="총대출액",
    aggfunc="sum"
)
print(pivot_result)

pivot_result["차이"] = abs(pivot_result[1] - pivot_result[2]).abs()
result = pivot_result["차이"].idxmax()
print(result)
```

```python
import pandas as pd

df = pd.read_csv("https://raw.githubusercontent.com/lovedlim/bigdata_analyst_cert_v2/main/part4/ch9/crime.csv")

cond1 = df["구분"] == "발생건수"
cond2 = df["구분"] == "검거건수"
df1 = df[cond1].iloc[:, 2:]
df2 = df[cond2].iloc[:, 2:]

df1 = df1.reset_index(drop=True)
df2 = df2.reset_index(drop=True)
df3 = df2 / df1

listbox = df3.idxmax(axis=1)

result = 0
for index, item in enumerate(listbox):
  result = result + df2.loc[index, item]
print(result)
```

```python
import pandas as pd

df = pd.read_csv("https://raw.githubusercontent.com/lovedlim/bigdata_analyst_cert_v2/main/part4/ch9/crime.csv")

df_long = pd.melt(df, id_vars=["연도", "구분"], var_name="범죄유형", value_name="건수")

df_pivot = df_long.pivot_table(
    index=["연도", "범죄유형"],
    columns="구분",
    values="건수"
).reset_index()

df_pivot["검거율"] = df_pivot["검거건수"] / df_pivot["발생건수"]

listbox = df_pivot.groupby("연도")["검거율"].idxmax()

max_crime = df_pivot.loc[listbox]
result = max_crime["검거건수"].sum()
print(result)
```

```python
import pandas as pd

df = pd.read_csv("https://raw.githubusercontent.com/lovedlim/bigdata_analyst_cert_v2/main/part4/ch9/hr.csv")

m = df["만족도"].mean()
df["만족도"] = df["만족도"].fillna(m)

gm = df.groupby(["부서", "성과등급"])["근속연수"].transform("mean")
gm = gm.astype(int)
df["근속연수"] = df["근속연수"].fillna(gm)

df["연봉_근속연수"] = df["연봉"] / df["근속연수"]
df_year = df.nlargest(3, "연봉_근속연수")
A = df_year.iloc[-1]["근속연수"]

df["연봉_만족도"] = df["연봉"] / df["만족도"]
df_like = df.nlargest(2, "연봉_만족도")
B = df_like.iloc[-1]["교육참가횟수"]

result = A + B
print(result)
```

```python
import pandas as pd

df = pd.read_csv("https://raw.githubusercontent.com/lovedlim/bigdata_analyst_cert_v2/main/part4/ch9/hr.csv")

m = df["만족도"].mean()
df["만족도"] = df["만족도"].fillna(m)

gm = df.groupby(["부서", "성과등급"])["근속연수"].transform("mean")
gm = gm.astype(int)
df["근속연수"] = df["근속연수"].fillna(gm)

df["연봉_근속연수"] = df["연봉"] / df["근속연수"]
df_year = df.sort_values(by="연봉_근속연수", ascending=False)
A = df_year.iloc[2]["근속연수"]

df["연봉_만족도"] = df["연봉"] / df["만족도"]
df_year = df.sort_values(by="연봉_만족도", ascending=False)
B = df_year.iloc[1]["교육참가횟수"]

result = A + B
print(result)
```

### 1. 데이터 재구성과 그룹화

데이터의 형태를 바꾸거나 그룹별 통계를 낼 때 사용합니다.

* **`melt()`**: '와이드(Wide)'한 데이터를 '롱(Long)'한 형태로 녹여서 재구성합니다. 여러 컬럼으로 나열된 변수들을 하나의 컬럼으로 합칠 때 유용합니다.
* **`groupby()`**: 특정 컬럼을 기준으로 데이터를 그룹으로 묶습니다. (예: 성별로 묶어서 평균 키 구하기)
* **`transform()`**: `groupby`와 함께 자주 쓰이며, 그룹별 통계치(평균 등)를 계산하여 **원래 데이터프레임의 행 길이에 맞춰** 값을 돌려줍니다.
* **`reset_index(drop=True)`**: 데이터 정렬이나 필터링 후 흐트러진 인덱스(번호)를 0부터 다시 순서대로 매깁니다. `drop=True`를 쓰면 기존의 복잡한 인덱스는 삭제합니다.
* **`sort_values()`**: 데이터를 특정 기준 컬럼 값에 따라 오름차순이나 내림차순으로 정렬합니다.

### 2. 데이터 스케일링: MinMaxScaler

데이터의 범위를 **0과 1 사이**로 압축하는 기술입니다.

#### **`fit_transform()`**
StandardScaler와 마찬가지로 학습(fit)과 변환(transform)을 동시에 합니다.


$$x_{scaled} = \frac{x - x_{min}}{x_{max} - x_{min}}$$

* **특징**: 모든 데이터가 정확히 0~1 사이에 위치하게 됩니다.
* **언제 쓰나요?**: 데이터의 분포가 정규분포가 아니거나, 모델이 0~1 사이의 입력을 선호할 때(예: 딥러닝, KNN 등) 사용합니다.

### 3. 한눈에 보는 비교표

| 메서드 | 핵심 역할 | 비유 |
| :--- | :--- | :--- |
| **`melt()`** | 열을 행으로 내리기 | 옆으로 긴 표를 아래로 길게 만들기 |
| **`groupby()`** | 데이터 집합 만들기 | 같은 팀끼리 모으기 |
| **`transform()`** | 그룹 통계치 뿌리기 | 팀 평균 점수를 각 팀원 옆에 적어주기 |
| **`MinMaxScaler`** | 0~1 사이 규격화 | 모든 점수를 100점 만점 환산 점수로 바꾸기 |
