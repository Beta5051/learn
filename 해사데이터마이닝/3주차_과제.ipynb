```python
import pandas as pd

df = pd.read_csv("https://raw.githubusercontent.com/lovedlim/bigdata_analyst_cert/main/part4/ch3/members.csv")

df_cntNull = df.isnull().sum()

df_cntNull = df_cntNull.sort_values(ascending=False)

print(df_cntNull.index[0])
```

```python
import pandas as pd

df = pd.read_csv("https://raw.githubusercontent.com/lovedlim/bigdata_analyst_cert/main/part4/ch7/student_assessment.csv")

df = df.dropna()

id = df['id_assessment'].value_counts().idxmax()
cond = df['id_assessment'] == id
df = df[cond]

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
df['score'] = scaler.fit_transform(df[['score']])
print(round(df['score'].max(), 3))
```

### 1. Pandas DataFrame 주요 메서드

데이터의 상태를 확인하고 형태를 바꾸는 데 사용됩니다.

* **`isnull()`**: 데이터프레임 내의 값이 결측치(NaN)인지 확인합니다. 보통 `.isnull().sum()` 형태로 써서 컬럼별 결측치 개수를 파악합니다.
* **`dropna()`**: 결측치가 포함된 행(row)이나 열(column)을 통째로 삭제합니다.
* **`fillna()`**: (지난번 설명처럼) 결측치를 특정 값(평균, 중앙값 등)으로 채웁니다.
* **`sort_values()`**: 특정 컬럼의 값을 기준으로 데이터를 정렬합니다. (오름차순/내림차순)
* **`value_counts()`**: 범주형 데이터에서 각 항목이 몇 개씩 있는지 개수를 세어줍니다.
* **`quantile(q)`**: 데이터를 크기순으로 세웠을 때 $q$ 위치에 있는 값을 찾습니다. (예: `0.5`는 중앙값, `0.25`는 1사분위수)
* **`max()`**: 가장 큰 값을 반환합니다.
* **`idxmax()`**: 가장 큰 값을 가진 데이터의 **인덱스(이름/위치)**를 반환합니다.

### 2. Python 기본 함수 (Built-in Functions)

파이썬 자체에서 제공하는 기초 함수입니다.

* **`abs()`**: 절댓값을 구합니다. (오차 계산 시 음수를 제거할 때 자주 씁니다.)
* **`len()`**: 데이터의 길이(개수)를 반환합니다. 데이터프레임에서는 전체 행의 개수를 알 수 있습니다.

### 3. 데이터 스케일링: StandardScaler

데이터 분석 전, 서로 다른 단위(예: 몸무게 kg과 키 cm)를 맞춰주는 **표준화** 작업에 쓰입니다.

#### **`fit_transform()`**
이 메서드는 두 가지 단계를 한 번에 수행합니다.
1.  **fit**: 데이터의 평균($\mu$)과 표준편차($\sigma$)를 계산하여 학습합니다.
2.  **transform**: 계산된 값을 바탕으로 데이터를 변환합니다.



$$z = \frac{x - \mu}{\sigma}$$

이 과정을 거치면 모든 데이터는 **평균이 0, 표준편차가 1**인 표준정규분포 형태로 변하게 되어, 머신러닝 모델이 데이터를 더 잘 학습할 수 있게 됩니다.
